```python
import requests
from bs4 import BeautifulSoup

# Fonction pour récupérer le contenu d'une page web
def get_page_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.text
        else:
            print(f"Erreur {response.status_code} lors de la récupération de la page {url}")
            return None
    except Exception as e:
        print(f"Erreur lors de la récupération de la page {url}: {str(e)}")
        return None
```

```python
# On essaye de webscrapper un résumé d'article parru sur le site europepmc, la page web résume un rapport scientifique parru par l'entreprise pfizer
# On cherche à webscrapper le résumé de ce rapport. Ensuite on pourra automatiser le processus pour webscrapper tout les résumés des rapports de pfizer
# Enfin on pourra automatiser une nouvelle fois, de manière à récupérer le résumé de parution de plusieurs entreprises

page_content = get_page_content('https://europepmc.org/article/PMC/PMC10677964')
page_content
```
